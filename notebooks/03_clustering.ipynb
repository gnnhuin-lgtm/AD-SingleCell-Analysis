# 03 — Dimensionality Reduction and Clustering

Perform PCA, UMAP, and Leiden/Louvain clustering.

```python
import scanpy as sc
import anndata as ad

from src.analysis import Clustering
from src.visualization import Visualizer
from src.utils import load_config

config = load_config('config/config.yaml')
adata = sc.read_h5ad('data/processed/02_normalized.h5ad')
```

```python
clustering = Clustering(adata, config=config)

# PCA
adata = clustering.pca()
sc.pl.pca_variance_ratio(adata, n_pcs=50)
sc.pl.pca(adata, color='total_counts')
```

```python
# UMAP
adata = clustering.umap()
sc.pl.umap(adata, color='total_counts')
```

```python
# Leiden clustering
adata = clustering.leiden(resolution=config['clustering']['resolution'])
sc.pl.umap(adata, color='leiden', legend_loc='on data', legend_fontsize=8)
```

```python
adata.write('data/processed/03_clustered.h5ad')
```
