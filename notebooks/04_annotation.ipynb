# 04 — Cell Type Annotation

Automatically annotate cell types using known brain marker genes.

```python
import scanpy as sc
import anndata as ad
import matplotlib.pyplot as plt

from src.annotation import CellTypeAnnotation
from src.utils import load_config

config = load_config('config/config.yaml')
adata = sc.read_h5ad('data/processed/03_clustered.h5ad')
```

```python
# Score cell types
annotator = CellTypeAnnotation(adata)
scores = annotator.score_cell_types()
print(scores.head())
```

```python
# Visualize
sc.pl.umap(adata, color='cell_type', legend_loc='on data', legend_fontsize=7)
sc.pl.dotplot(adata, var_names=['GFAP', 'SYT1', 'TREM2', 'MBP', 'PDGFRA'],
              groupby='cell_type')
```
