# 02 — Data Preprocessing

Normalize data, select highly variable genes, and prepare for downstream analysis.

```python
import scanpy as sc
import anndata as ad

from src.preprocessing import Normalize, VariableGeneSelection
from src.utils import setup_logging, load_config

setup_logging()
config = load_config('config/config.yaml')
adata = sc.read_h5ad('data/processed/01_qced.h5ad')
```

```python
# Normalize and log-transform
normalizer = Normalize(adata, config=config)
adata = normalizer.normalize_and_log()
```

```python
# Select highly variable genes
hvg = VariableGeneSelection(adata, config=config)
adata = hvg.select_hvg(
    n_top_genes=config['hvg']['n_top_genes'],
    flavor=config['hvg']['flavor'],
)
sc.pl.highly_variable_genes(adata)
```

```python
# Subset to HVGs
adata = adata[:, adata.var['highly_variable']].copy()
print(f'After HVG selection: {adata.n_obs} cells x {adata.n_vars} genes')
```

```python
adata.write('data/processed/02_normalized.h5ad')
```
